```python
import pandas as pd
india_weather=pd .DataFrame({
    "city":["mumbai","chennai","kolkata"],
    "temprature":[32,34,29.5],
    "humidity":[60,45,78]
    
})
us_weather=pd .DataFrame({
    "city":["new-york","chicago","orlando"],
    "temprature":[34,33.7,19.5],
    "humidity":[55,65,48]
    
})
```

```python

us_weather.head()
```

```python
india_weather.head()
```

# CONCATINATION BY COLUMNS


```python
df3=pd.concat([us_weather,india_weather])
```

```python
df3
```

```python
df3=pd.concat([us_weather,india_weather],ignore_index=True)
```

```python
df3
```

# CONCATINATION BY ROWS


```python
df4=pd.concat([india_weather,us_weather],axis=1)
```

```python
df4
```

# MERGING

```python
table_1=pd.DataFrame({
    "city":["mumbai","kolkata","chennai"],
    "temprature":[30,35,39]
})
table_2=pd.DataFrame({
    "city":["mumbai","hyderabad","chennai"],
    "humidity":[30,34,39]
})
df5=pd.merge(table_1,table_2,on="city")
```

```python
df5.head()
```

```python
df5=pd.merge(table_1,table_2,on="city",how="outer")
```

```python
df5
```
